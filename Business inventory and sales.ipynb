```python
df = pd.read_csv("TOTBUSIMNSA.csv")
df.head()
```

```python
# plots of values in businv
fig, axes = plt.subplots(2, 1, figsize=(6, 12))
fig.tight_layout()

# Plot the histogram
sns.histplot(df['value'], ax=axes[0], bins=10)
axes[0].set_title('value')


# Plot the box plot
sns.boxplot(data=df, x='value', ax=axes[1])
axes[1].set_title('Box Plot of value')
```

```python
df.plot(kind='scatter', x='realtime_end', y='realtime_start', alpha=0.5)
plt.title('realtime_end vs realtime_start')
plt.ylabel('realtime_start')
plt.xlabel('realtime_end')
plt.show()
```

```python
#new data on superstore sales
df = pd.read_csv("Superstore.csv", encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
summary_stats = df[['Sales','Quantity','Profit']].describe()
summary_stats
```

```python
#plots on sales and quantity
fig, axes = plt.subplots(2, 3, figsize=(25, 19))
fig.subplots_adjust(hspace=0.5)

# Plot the histogram for Sales
sns.histplot(df['Sales'], ax=axes[0, 0], bins=10)
axes[0, 0].set_title('Sales')

# Plot the box plot for Sales
sns.boxplot(data=df, x='Sales', ax=axes[0, 1])
axes[0, 1].set_title('Box Plot of Sales')

# Plot the KDE for Sales
sns.kdeplot(data=df, x='Sales', fill=True, ax=axes[0, 2])
axes[0, 2].set_title('KDE of Sales')

# Plot the histogram for Quantity
sns.histplot(df['Quantity'], ax=axes[1, 0], bins=10, color='green')
axes[1, 0].set_title('Quantity')

# Plot the box plot for Quantity
sns.boxplot(data=df, x='Quantity', ax=axes[1, 1], color='green')
axes[1, 1].set_title('Box Plot of Quantity')
```

```python
df.plot(x='Profit', y='Sales')
 
plt.xlabel('Profit')
plt.ylabel('Sales')
plt.title('Sales vs Profit')

plt.show()
```
