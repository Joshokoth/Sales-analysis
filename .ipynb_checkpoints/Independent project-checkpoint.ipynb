```python
#1. sales distribution visualization
# Create a figure with three subplots
fig, axes = plt.subplots(3, 1, figsize=(6, 12))
fig.tight_layout()

# Plot the histogram
sns.histplot(df['SalePrice'], ax=axes[0], bins=10)
axes[0].set_title('SalePrice Distribution')


# Plot the box plot
sns.boxplot(data=df, x='SalePrice', ax=axes[1])
axes[1].set_title('Box Plot of SalePrice Distribution')


# Plot the KDE
sns.kdeplot(data=df, x='SalePrice', fill=True, ax=axes[2])
axes[2].set_title('KDE of SalePrice Distribution')

# Set common labels
for ax in axes:
    ax.set_xlabel('SalePrice')
    ax.set_ylabel('Quantity')

plt.show()
```
